# **Data_Preprocessing**

```python
main_df.isnull().sum()
```

```python
main_df.dropna(subset = ["RainTomorrow"], inplace=True)
```

```python
main_df.isnull().sum()
```

```python
date = pd.to_datetime(main_df['Date'])
year = date.dt.year
```

```python
train_df = main_df[year < 2015]
val_df = main_df[year == 2015]
test_df = main_df[year > 2015]
```

```python
input_cols = train_df.columns[1:-1].tolist()
target_col = 'RainTomorrow'
```

```python
train_inputs = train_df[input_cols].copy()
train_targets = train_df[target_col].copy()
```

```python
val_inputs = val_df[input_cols].copy()
val_targets = val_df[target_col].copy()
```

```python
test_inputs = test_df[input_cols].copy()
test_targets = test_df[target_col].copy()
```

```python
numeric_cols = train_inputs.select_dtypes(include=np.number).columns.tolist()
categoric_cols = train_inputs.select_dtypes(exclude=np.number).columns.tolist()
```

```python
train_inputs[numeric_cols].head()
```

```python
train_inputs[categoric_cols].head()
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='median')
imputer.fit(train_inputs[numeric_cols]) #only train on testsets.
```

```python
train_inputs[numeric_cols] = imputer.transform(train_inputs[numeric_cols])
val_inputs[numeric_cols] = imputer.transform(val_inputs[numeric_cols])
test_inputs[numeric_cols] = imputer.transform(test_inputs[numeric_cols])
```

```python
train_inputs[numeric_cols].isna().sum()
```

```python
train_inputs[categoric_cols].isna().sum()
```

```python
from sklearn.preprocessing import MinMaxScaler
scalar = MinMaxScaler()
scalar.fit(train_inputs[numeric_cols])
```

```python
train_inputs[numeric_cols] = scalar.transform(train_inputs[numeric_cols])
val_inputs[numeric_cols] = scalar.transform(val_inputs[numeric_cols])
test_inputs[numeric_cols] = scalar.transform(test_inputs[numeric_cols])
```

```python
train_inputs[numeric_cols].head()
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
encoder.fit(train_inputs[categoric_cols])
```

```python
encoded_cols = list(encoder.get_feature_names_out(categoric_cols))
```

```python
encoded_cols
```

```python
train_inputs[encoded_cols] = encoder.transform(train_inputs[categoric_cols])
val_inputs[encoded_cols] = encoder.transform(val_inputs[categoric_cols])
test_inputs[encoded_cols] = encoder.transform(test_inputs[categoric_cols])
```

```python
train_X = train_inputs[numeric_cols + encoded_cols]
val_X = val_inputs[numeric_cols + encoded_cols]
test_X = test_inputs[numeric_cols + encoded_cols]
```
